```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata

# Retrieve your Google API key securely from Colab secrets
os.environ["GOOGLE_API_KEY"] = userdata.get("HK_GEM_KEY_003") # <--- ENSURE 'GOOGLE_API_KEY' MATCHES THE NAME IN COLAB SECRETS

# Create a Gemini Pro model
gemini_llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    temperature=0.2  # adjust for creativity or conservatism
)

print("Gemini Pro LLM ready!")
```

```python
def inspect_python_error(error_msg: str) -> str:
    """
    A simple Python error inspection helper.
    """
    if "ZeroDivisionError" in error_msg:
        return "It appears to be a division by zero issue. Check your divisor logic."
    elif "IndexError" in error_msg:
        return "IndexError: check if your list or array indices go out of range."
    return "Hmm, can’t confidently diagnose from that message alone."
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model=gemini_llm,
    tools=[inspect_python_error],
    system_prompt="You are a Python debugging assistant using Google Gemini Pro."
)

print("Agent initialized with Gemini Pro!")
```

```python
user_input = "Here’s the error: IndexError: list index out of range"
output = agent.invoke(user_input)
print(output)
```
